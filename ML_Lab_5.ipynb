(A1) This code fits a straight‑line model y = w_0 + w_1 x to the data in column 0 (feature) and column 1 (target).
```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'df' is your DataFrame loaded earlier (e.g., pd.read_csv('data.csv'))
# Using column 0 as feature (X) and column 1 as target (y)
X = df[[0]].values  # Feature: must be 2D array for sklearn
y = df[1].values    # Target: 1D array

# Split data: 70% training, 30% testing (random_state ensures reproducibility)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train linear regression model on training data
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on both training and test sets
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Evaluate model performance
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
print(f"Training R² Score: {train_r2:.3f}")
print(f"Testing R² Score: {test_r2:.3f}")

# Visualize training data with regression line
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='green', alpha=0.6, label='Training Data')
plt.plot(X_train, y_train_pred, color='blue', linewidth=2.5, label='Fitted Line')
plt.xlabel("Feature (Column 0)")
plt.ylabel("Target (Column 1)")
plt.title("Linear Regression: Training Data Fit")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

(A2) This code computes 4 essential metrics (MSE, RMSE, MAPE, R²) for both training and test sets to assess our linear regression model's performance. R² closest to 1.0 indicates good fit; low RMSE/MAPE shows accurate predictions. Training metrics typically look better than test metrics—watch for large gaps indicating overfitting!

```python
import pandas as pd                                    # For loading Excel/CSV data
from sklearn.linear_model import LinearRegression     # Linear regression model
from sklearn.model_selection import train_test_split  # Data splitting utility
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score  # Evaluation metrics
import matplotlib.pyplot as plt                       # Plotting library
import numpy as np                                    # Math operations (sqrt)


df = pd.read_excel('BERT_Embeddings.xlsx')
print(f" Data loaded! Shape: {df.shape}")           # Confirm data loaded


X = df[[0]].values                                    # Feature: Column 0 (must be 2D array)
y = df[1].values                                      # Target: Column 1 (1D array)
print(f"Features shape: {X.shape}, Target shape: {y.shape}")


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"Training samples: {len(X_train)}, Test samples: {len(X_test)}")

reg = LinearRegression()                              # Create regression model object
reg.fit(X_train, y_train)                             # Train model on training data
print(" Model trained successfully!")

def calculate_model_performance(model, X_train, y_train, X_test, y_test):
    """Calculate 8 key metrics for model performance"""
    train_pred = model.predict(X_train)                 # Training predictions
    test_pred = model.predict(X_test)                   # Test predictions

    # Training metrics
    train_mse = mean_squared_error(y_train, train_pred)  # Mean Squared Error
    train_rmse = np.sqrt(train_mse)                      # Root Mean Squared Error
    train_mape = mean_absolute_percentage_error(y_train, train_pred)  # MAPE
    train_r2 = r2_score(y_train, train_pred)             # R² score

    # Test metrics
    test_mse = mean_squared_error(y_test, test_pred)     # Test MSE
    test_rmse = np.sqrt(test_mse)                        # Test RMSE
    test_mape = mean_absolute_percentage_error(y_test, test_pred)     # Test MAPE
    test_r2 = r2_score(y_test, test_pred)                # Test R²

    return (train_mse, train_rmse, train_mape, train_r2,
            test_mse, test_rmse, test_mape, test_r2)

metrics = calculate_model_performance(reg, X_train, y_train, X_test, y_test)
train_mse, train_rmse, train_mape, train_r2, test_mse, test_rmse, test_mape, test_r2 = metrics

print("\n TRAINING SET METRICS:")
print(f"MSE:  {train_mse:.4f}")
print(f"RMSE: {train_rmse:.4f}")
print(f"MAPE: {train_mape:.4f}")
print(f"R²:   {train_r2:.4f}")

print("\n TEST SET METRICS:")
print(f"MSE:  {test_mse:.4f}")
print(f"RMSE: {test_rmse:.4f}")
print(f"MAPE: {test_mape:.4f}")
print(f"R²:   {test_r2:.4f}")

plt.figure(figsize=(10, 6))                          # Set figure size
plt.scatter(X_train, y_train, color='blue', alpha=0.6, label='Training Data')  # Scatter plot of training points
plt.plot(X_train, reg.predict(X_train), color='red', linewidth=2.5, label='Regression Line')  # Fitted line
plt.xlabel("Feature (Column 0)")                      # X-axis label
plt.ylabel("Target (Column 1)")                       # Y-axis label
plt.title("Linear Regression: Training Data Fit")     # Plot title
plt.legend()                                          # Show legend
plt.grid(True, alpha=0.3)                             # Add light grid
plt.tight_layout()                                    # Adjust layout
plt.show()                                            # Display plot

print("\n COMPLETE ANALYSIS DONE!")
```

(A3) Predicts column 1 using ALL ~767 numeric columns. Massive jump from simple (1 feature) to multiple linear regression (767 features). Evaluates train/test performance with 4 metrics each.

Key upgrade: A2 used just column 0 to column 1. Now uses 767x more attributes for much richer prediction power

```python
from sklearn.linear_model import LinearRegression              # Import LinearRegression class for model training
from sklearn.model_selection import train_test_split          # Import train_test_split for data splitting
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score  # Import 3 evaluation metrics
import numpy as np                                            # Import NumPy for square root calculations

def prepare_data_all_features(df, target_col_name):            # Define function to prepare features from all numeric columns
    # Find all numeric columns numbered 0-767 (BERT embedding style)
    numeric_cols = [col for col in df.columns if isinstance(col, int) and 0 <= col <= 767]  # Filter only integer columns 0-767

    # Select all numeric columns EXCEPT the target column as features
    feature_cols = [col for col in numeric_cols if col != target_col_name]  # Exclude target from features
    X = df[feature_cols].values                                  # Convert feature columns to numpy array (samples × 767 features)

    # Extract target variable as 1D array
    y = df[target_col_name].values                               # Get target column values

    # Split data: 70% training, 30% testing with fixed random seed for reproducibility
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)  # Create train/test splits
    return X_train, X_test, y_train, y_test                      # Return all 4 split datasets

def train_linear_model(X_train, y_train):                      # Define function to train linear regression model
    model = LinearRegression()                                  # Create new LinearRegression instance
    model.fit(X_train, y_train)                                 # Train model using training features and targets
    return model                                                # Return trained model

def evaluate_model_metrics(model, X_train, y_train, X_test, y_test):  # Define function to calculate 8 performance metrics
    y_train_pred = model.predict(X_train)                       # Generate predictions for training set
    y_test_pred = model.predict(X_test)                         # Generate predictions for test set

    # Training Set Metrics calculations
    mse_train = mean_squared_error(y_train, y_train_pred)       # Calculate Mean Squared Error for training
    rmse_train = np.sqrt(mse_train)                             # Calculate RMSE (sqrt of MSE) for training
    mape_train = mean_absolute_percentage_error(y_train, y_train_pred)  # Calculate MAPE for training
    r2_train = r2_score(y_train, y_train_pred)                  # Calculate R² score for training

    # Test Set Metrics calculations
    mse_test = mean_squared_error(y_test, y_test_pred)          # Calculate Mean Squared Error for test set
    rmse_test = np.sqrt(mse_test)                               # Calculate RMSE for test set
    mape_test = mean_absolute_percentage_error(y_test, y_test_pred)     # Calculate MAPE for test set
    r2_test = r2_score(y_test, y_test_pred)                     # Calculate R² score for test set

    # Return all 8 metrics in exact same order
    return mse_train, rmse_train, mape_train, r2_train, mse_test, rmse_test, mape_test, r2_test  # Return metrics tuple


target_col_index = 1                                          # Set target column index (column 1 as target)

# Prepare training/testing data using ALL other numeric columns as features
X_train, X_test, y_train, y_test = prepare_data_all_features(df, target_col_index)  # Call data preparation function

# Train linear regression model
model = train_linear_model(X_train, y_train)                  # Train model and store in 'model' variable

# Calculate performance metrics
mse_train, rmse_train, mape_train, r2_train, mse_test, rmse_test, mape_test, r2_test = evaluate_model_metrics(  # Extract 8 metrics
    model, X_train, y_train, X_test, y_test                   # Pass trained model + all data splits
)

# Display TRAINING results with original formatting
print("Training Set Metrics:")                             # Print training section header
print(f"MSE: {mse_train:.4f}")                               # Print training MSE (4 decimals)
print(f"RMSE: {rmse_train:.4f}")                             # Print training RMSE (4 decimals)
print(f"MAPE: {mape_train:.4f}")                             # Print training MAPE (4 decimals)
print(f"R²: {r2_train:.4f}")                                 # Print training R² (4 decimals)

# Display TEST results with original formatting
print("\nTest Set Metrics:")                               # Print test section header (with newline)
print(f"MSE: {mse_test:.4f}")                                # Print test MSE (4 decimals)
print(f"RMSE: {rmse_test:.4f}")                              # Print test RMSE (4 decimals)
print(f"MAPE: {mape_test:.4f}")                              # Print test MAPE (4 decimals)
print(f"R²: {r2_test:.4f}")                                  # Print test R² (4 decimals)
```

(A4) Clusters high-dimensional data points into 2 groups using K-Means on ~767 numeric features. Labels show which cluster (0/1) each sample belongs to; centers are the 2 cluster centroids in 767D space. Perfect companion to regression—reveals natural groupings missed by linear models

```python
from sklearn.cluster import KMeans                                      # Import KMeans clustering algorithm

def prepare_clustering_data(df, target_col_index):                      # Define function to prepare clustering features
    # Detect all integer-numbered columns (0-767 range) suitable for clustering
    numeric_cols = [col for col in df.columns if isinstance(col, int) and 0 <= col <= 767]  # Filter numeric columns 0-767

    # Remove target column from clustering features (we cluster on other dimensions)
    features_for_clustering = [col for col in numeric_cols if col != target_col_index]  # Exclude target column

    # Return numpy array of clustering features (samples × features)
    return df[features_for_clustering].values                           # Convert selected columns to numpy array

def perform_kmeans_clustering(X, n_clusters=2, random_state=0):          # Define KMeans clustering function
    # Initialize KMeans with specified clusters and reproducible random seed
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state, n_init="auto")  # Create KMeans model (auto init)

    # Fit clustering model to the high-dimensional data
    kmeans.fit(X)                                                       # Train KMeans on feature matrix

    # Return cluster assignments and centroid coordinates
    return kmeans.labels_, kmeans.cluster_centers_                      # Return labels array + centers matrix


target_col_index = 1                                                    # Set target column to exclude (column 1)

# Prepare high-dimensional features for clustering (excludes target column 1)
X_cluster = prepare_clustering_data(df, target_col_index)               # Get clustering-ready data matrix

# Execute K-Means with 2 clusters on ~767-dimensional embeddings
labels, centers = perform_kmeans_clustering(X_cluster, n_clusters=2, random_state=0)  # Run clustering

# Display clustering results
print("Cluster Labels (first 10):", labels[:10])                        # Show first 10 cluster assignments (0 or 1)
print("Cluster Centers Shape:", centers.shape)                          # Show shape of 2 centroids × 767 features
```

(A5) Evaluates K-Means clustering quality on ~767D embeddings using 3 standard metrics. Silhouette increses measures separation vs cohesion; Calinski-Harabasz increses measures variance ratio; Davies-Bouldin decreases measures cluster similarity.

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score  # Import 3 key clustering quality metrics
from sklearn.cluster import KMeans                                                   # Import KMeans (backup access to labels if needed)

def evaluate_clustering_metrics(X, labels):                                          # Define function to compute clustering quality scores
    silhouette = silhouette_score(X, labels)                                         # Calculate Silhouette Score (-1 to +1, higher = better cluster separation)
    ch_score = calinski_harabasz_score(X, labels)                                    # Calculate Calinski-Harabasz Score (higher = better)
    db_index = davies_bouldin_score(X, labels)                                       # Calculate Davies-Bouldin Index (lower = better)
    return silhouette, ch_score, db_index                                            # Return all 3 metrics
# Uses X_cluster and labels from previous KMeans clustering

silhouette_score_val, ch_score_val, db_index_val = evaluate_clustering_metrics(X_cluster, labels)  # Compute metrics on clustered data

# Display clustering quality results with original formatting
print("Clustering Evaluation Metrics:")                                           # Print section header
print(f"Silhouette Score: {silhouette_score_val:.4f}")                              # Print Silhouette (↑ better, 0.7+ = excellent)
print(f"Calinski-Harabasz Score: {ch_score_val:.4f}")                               # Print CH Score (↑ better, measures variance ratio)
print(f"Davies-Bouldin Index: {db_index_val:.4f}")                                  # Print DB Index (↓ better, <1 = very good)
```

(A6) Tests K=2-10 clusters on ~767D embeddings, plots Silhouette/CH/DB scores to find optimal k. Silhouette peaks + CH peaks + DB dips = best k. Visual "elbow" reveals natural cluster count in high-dimensional data

```python
from sklearn.cluster import KMeans                                        # Import KMeans clustering algorithm
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score  # Import 3 clustering metrics
import matplotlib.pyplot as plt                                           # Import plotting library

def evaluate_kmeans_for_k_range(X, k_values, random_state=42):            # Define function to test multiple k values
    silhouette_scores = []                                                # Initialize empty list for silhouette scores
    ch_scores = []                                                        # Initialize empty list for CH scores
    db_scores = []                                                        # Initialize empty list for DB scores

    # Loop through each k value to test
    for k in k_values:                                                    # Iterate over range of cluster counts
        kmeans = KMeans(n_clusters=k, random_state=random_state, n_init="auto")  # Create KMeans model for current k
        labels = kmeans.fit_predict(X)                                    # Fit model and get cluster labels

        # Calculate all 3 metrics for current k
        silhouette = silhouette_score(X, labels)                          # Compute Silhouette Score for this k
        ch = calinski_harabasz_score(X, labels)                           # Compute Calinski-Harabasz Score
        db = davies_bouldin_score(X, labels)                              # Compute Davies-Bouldin Index

        # Store metrics for plotting
        silhouette_scores.append(silhouette)                              # Add silhouette score to list
        ch_scores.append(ch)                                              # Add CH score to list
        db_scores.append(db)                                              # Add DB score to list

    return silhouette_scores, ch_scores, db_scores                        # Return all 3 score lists

def plot_clustering_metrics(k_values, silhouette_scores, ch_scores, db_scores):  # Define plotting function
    plt.figure(figsize=(15, 4))                                           # Create figure with 3 subplots side-by-side

    # Silhouette Score plot (first subplot)
    plt.subplot(1, 3, 1)                                                  # Select 1st subplot (1 row, 3 cols, position 1)
    plt.plot(k_values, silhouette_scores, marker='o')                     # Plot silhouette scores with circle markers
    plt.title("Silhouette Score")                                         # Set subplot title
    plt.xlabel("No of Clusters (k)")                                  # Set x-axis label
    plt.ylabel("Score")                                                   # Set y-axis label

    # Calinski-Harabasz Score plot (second subplot)
    plt.subplot(1, 3, 2)                                                  # Select 2nd subplot
    plt.plot(k_values, ch_scores, marker='*', color='orange')             # Plot CH scores (green line)
    plt.title("Calinski-Harabasz Score")                                  # Set subplot title
    plt.xlabel("N0 of Clusters (k)")                                  # Set x-axis label
    plt.ylabel("Score")                                                   # Set y-axis label

    # Davies-Bouldin Index plot (third subplot)
    plt.subplot(1, 3, 3)                                                  # Select 3rd subplot
    plt.plot(k_values, db_scores, marker='o', color='brown')               # Plot DB scores (red line)
    plt.title("Davies-Bouldin Index")                                     # Set subplot title
    plt.xlabel("N0 of Clusters (k)")                                  # Set x-axis label
    plt.ylabel("Score")                                 # Set y-axis label (note: lower=better)

    plt.tight_layout()                                                    # Adjust subplot spacing
    plt.show()                                                            # Display all 3 plots


def prepare_clustering_data(df, target_col_index):                      # Redefine clustering data prep function
    numeric_cols = [col for col in df.columns if isinstance(col, int) and 0 <= col <= 767]  # Find numeric columns 0-767
    features_for_clustering = [col for col in numeric_cols if col != target_col_index]     # Exclude target column
    return df[features_for_clustering].values                            # Return feature matrix

target_col_index = 1                                                   # Set target column to exclude from clustering
X_cluster = prepare_clustering_data(df, target_col_index)              # Prepare clustering data (~767 features)

# Test k=2 through k=10
k_values = list(range(2, 11))                                          # Create list [2,3,4,5,6,7,8,9,10]

# Evaluate clustering quality across k range
silhouette_scores, ch_scores, db_scores = evaluate_kmeans_for_k_range(X_cluster, k_values)  # Get metrics for all k values

# Generate elbow plots for optimal k selection
plot_clustering_metrics(k_values, silhouette_scores, ch_scores, db_scores)  # Create 3-panel visualization
```

(A7) Elbow method plots inertia (WCSS) vs k=2-19 on ~767D data. "Elbow" point (sharp bend where slope flattens) = optimal cluster count. Complements previous silhouette analysis

```python
from sklearn.cluster import KMeans                                        # Import KMeans clustering algorithm
import matplotlib.pyplot as plt                                           # Import matplotlib for plotting

def compute_elbow_distortions(X, k_range, random_state=42):               # Define function to calculate inertia for k range
    distortions = []                                                      # Initialize empty list to store distortion values

    # Test each k value in range
    for k in k_range:                                                     # Loop through each k value
        kmeans = KMeans(n_clusters=k, random_state=random_state, n_init="auto")  # Create KMeans model for current k
        kmeans.fit(X)                                                     # Fit KMeans to data
        distortions.append(kmeans.inertia_)                               # Store within-cluster sum of squares (WCSS)

    return distortions                                                     # Return list of distortion values

def plot_elbow_curve(k_range, distortions):                               # Define function to plot elbow curve
    plt.figure(figsize=(8, 5))                                            # Create figure with 8x5 size
    plt.plot(k_range, distortions, marker='*',color='darkred')                            # Plot k vs distortion with circle markers
    plt.title("Elbow Method for Optimal k")                               # Set plot title
    plt.xlabel("No of Clusters (k)")                                  # Set x-axis label
    plt.ylabel("Inertia (Distortion)")                                    # Set y-axis label (WCSS)
    plt.grid(True)                                                        # Add grid for readability
    plt.show()                                                            # Display the plot

def prepare_clustering_data(df, target_col_index):                      # Redefine data preparation function
    numeric_cols = [col for col in df.columns if isinstance(col, int) and 0 <= col <= 767]  # Find numeric columns 0-767
    features_for_clustering = [col for col in numeric_cols if col != target_col_index]     # Exclude target column
    return df[features_for_clustering].values                            # Return feature matrix for clustering

target_col_index = 1                                                   # Set column 1 as target to exclude
X_cluster = prepare_clustering_data(df, target_col_index)              # Prepare ~767D clustering data

# Test k=2 to k=19
k_range = range(2, 20)                                                 # Create range [2,3,...,19]

# Calculate inertia (WCSS) for each k
distortions = compute_elbow_distortions(X_cluster, k_range)            # Get distortion values for plotting

# Generate elbow plot
plot_elbow_curve(k_range, distortions)                                 # Create and show elbow curve
```
