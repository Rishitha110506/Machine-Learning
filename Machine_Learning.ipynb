```python
##1. Write a program to count the number of vowels and consonants present in an input string.

text = input("Enter the string:")
vowels = "aeiouAEIOU"
vowel = 0
consonant = 0
for char in text:
  if char.isalpha():
    if char in vowels:
      vowel += 1
    else:
      consonant += 1

print("no of vowels in string:",vowel)
print("no of consonantsn in string:",consonant)
```

```python
##2. Write a program that accepts two matrices A and B as input and returns their product AB. Check if A & B are multipliable; if not, return error message.

rowsA = int(input("enter the no of rows in A:"))
colsA = int(input("enter the no of columns in A:"))
A = []
print("enter the elements of matrix A row by row:")
for _ in range(rowsA):
  row = list(map(int,input().split()))
  A.append(row)

rowsB = int(input("enter the no of rows in B:"))
colsB = int(input("enter the no of columns in B:"))
B = []
print("enter the elements of matrix B row by row:")
for _ in range(rowsB):
  row = list(map(int,input().split()))
  B.append(row)

if colsA != rowsB:
  print("error, matrices cant be multiplied")
else:
  result = [[0 for _ in range(colsB)] for _ in range(rowsA)]
  for i in range(rowsA):
    for j in range(colsB):
      for k in range(colsA):
        result[i][j] += A[i][k] * B[k][j]
  print("product of matrices A and B:")
  for row in result:
    print(row)


```

```python
##3. Write a program to find the number of common elements between two lists. The lists contain integers.

list1 = list(map(int,input("Enter the elements for list 1: ").split()))
list2 = list(map(int,input("Enter the elements for list 2: ").split()))

common = []
for x in list1:
  if x in list2 and x not in common:
    common.append(x)
print("print the common elements:",common)
print("no of common elements:",len(common))
```

```python
##4. Write a program that accepts a matrix as input and returns its transpose.

rows = int(input("enter no of rows:"))
cols = int(input("enter no of columns:"))

matrix = []
print("enter matrix row by row:")
for i in range(rows):
  matrix.append(list(map(int,input().split())))

transpose = []
for i in range(cols):
  new_row = []
  for j in range(rows):
    new_row.append(matrix[j][i])
  transpose.append(new_row)
print("transpose:")
for row in transpose:
  print(row)

```

```python
##5. Generate a list of 100 random numbers between 100 and 150. Find the mean, median and mode for these numbers.

import random
nums = [random.randint(100,150) for _ in range(100)]
print("Numbers:",nums)
n=len(nums)
#to find mean
mean = sum(nums)/n
#to find median
nums.sort()
if n % 2 == 0:
  median = (nums[n//2-1] + nums[n//2])/2
else:
  median = nums[n//2]
#to find mode
mode = nums[0]
maxCount = 0
for x in nums:
  count = nums.count(x)
  if count > maxCount:
    maxCount = count
    mode = x
print("Mean:",mean)
print("Median:",median)
print("Mode:",mode)
```
